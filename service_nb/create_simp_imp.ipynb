# Импорты

```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer
import pickle
```

# Загрузка чищенного датасета, разбивка по колонкам

```python
df = pd.read_csv(
    filepath_or_buffer='../data/cleared_df.csv',
    sep=';',
    index_col='SkillFactory_Id'
)
banks = [
    'BankA_decision',
    'BankB_decision',
    'BankC_decision',
    'BankD_decision',
    'BankE_decision'
]
df = df.drop(banks, axis='columns')
df = df.dropna(how='all')
df.info()
```

```python
df['Loan_term'].unique()
```

```python
# Список столбцов для заполнения "most_frequnces"
CAT_COLUMNS = [
    'education',
    'employment status',
    'Value',
    'Position',
    'Gender',
    'Family status',
    'ChildCount',
    'SNILS',
    'Merch_code',
    'Loan_term',
    'Goods_category',
]
# Список столбцов для заполнения "mean"
REG_COLUMNS = [
    'MonthProfit',
    'MonthExpense',
    'Loan_amount',
]
```

```python
df[CAT_COLUMNS].head(3)
```

```python
df[REG_COLUMNS].head(3)
```

# Создаем предобученные модели простых заполнителей SKlearn

```python
cat_imputer = SimpleImputer(strategy='most_frequent')
cat_imputer = cat_imputer.fit(df[CAT_COLUMNS])
reg_imputer = SimpleImputer(strategy='mean')
reg_imputer = reg_imputer.fit(df[REG_COLUMNS])
```

```python
da = df.copy()
da[CAT_COLUMNS] = cat_imputer.transform(df[CAT_COLUMNS])
da[REG_COLUMNS] = reg_imputer.transform(df[REG_COLUMNS])
da.head(3)
```

# Сохраняем модели простых заполнятелей

```python
pickle.dump(cat_imputer, open('../models/simple.cat', 'wb'))
pickle.dump(reg_imputer, open('../models/simple.reg', 'wb'))
```

# Проверка работоспособности

```python
cat_imputer = pickle.load(open('../models/simple.cat', 'rb'))
reg_imputer = pickle.load(open('../models/simple.reg', 'rb'))
```
